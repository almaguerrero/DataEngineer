## Exploratory Data Analysis

### - Reading the Data

```python
data = pd.read_csv('data.csv')

print(data.shape)
data.head()
```

```python
data.columns
```

### - Data Exploration

```python
data.describe()
```

```python
dt_1=data["LMSID_A"]
dt_2=data["DEAL"]
dt_3=data["DEALTYPE"]
dt_4=data["PROPTYPE_M"]
dt_5=data["PROPTYPE_J"]
dt_6=data["STATE"]
dt_7=data["LIQTYPE_JB"]
dt_8=data["LIQ_DT"]
dt_9=data["NTRAN"]
dt_10=data["DTRAN"]

df = pd.concat([dt_1,dt_2,dt_3,dt_4,dt_5,dt_6,dt_7,dt_8,dt_10,dt_9], axis=1)
df['DTRAN'] = pd.to_datetime(df["DTRAN"]).dt.strftime('%Y-%m-%d')
```

```python
plt.figure(figsize=(10,10))  ##tamaÃ±o de la figura 
sns.lineplot(x=df['DTRAN'], y=df['NTRAN'])
```

```python
plt.figure(figsize=(10,5))
chart = sns.stripplot(x=data['STATE'],y=data['NTRAN'],palette='rainbow')
chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
```
